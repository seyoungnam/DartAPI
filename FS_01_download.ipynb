```python
# 인터넷 상의 자료 불러오기 기초
import requests
```

```python
# url은 크롬 f12 누른 후 Headers->General->Request url
url = "http://dart.fss.or.kr/pdf/download/excel.do?rcp_no=20191114001273&dcm_no=6958001&lang=ko"
```

```python
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36"
```

```python
resp = requests.get(url, headers={"user-agent": user_agent})
```

```python
resp
```

```python
resp.content
```

```python
from io import BytesIO
import pandas as pd
```

```python
table = BytesIO(resp.content)
```

```python
pd.read_excel(table, sheet_name=0)
```

```python
pd.read_excel(table, sheet_name="기본정보")
```

```python
# 필요한 sheet의 내용만 편집하여 csv로 저장
data = pd.read_excel(table, sheet_name="연결 재무상태표", skiprows=5)
data.to_csv("연결재무상태표.csv", encoding="euc-kr")
```

```python
data
```

```python
# 여러개의 sheet를 각각의 csv파일로 하드디스크에 저장하기
pocket = ["연결 재무상태표", "연결 손익계산서", "연결 포괄손익계산서", "연결 자본변동표", "연결 현금흐름표"]
for sheet in pocket:
    data = pd.read_excel(table, sheet_name=sheet, skiprows=5)
    data.to_csv(sheet+".csv", encoding="euc-kr")
```
