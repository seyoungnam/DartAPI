```python
# (응용) DataFrame으로 전환
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

```python
isu_dcrs_de_ls = []
isu_dcrs_stle_ls = []
isu_dcrs_stock_knd_ls = []
isu_dcrs_qy_ls = []
isu_dcrs_mstvdv_fval_amount_ls = []
isu_dcrs_mstvdv_amount_ls = []

dict = json.loads(webpage)
for val in dict:
    if val == 'list':
        for ele in dict[val]:
            for sub_ele in ele:
                if sub_ele == 'isu_dcrs_de': isu_dcrs_de_ls.append(ele[sub_ele])
                elif sub_ele == 'isu_dcrs_stle': isu_dcrs_stle_ls.append(ele[sub_ele])
                elif sub_ele == 'isu_dcrs_stock_knd': isu_dcrs_stock_knd_ls.append(ele[sub_ele])
                elif sub_ele == 'isu_dcrs_qy': isu_dcrs_qy_ls.append(ele[sub_ele])
                elif sub_ele == 'isu_dcrs_mstvdv_fval_amount': isu_dcrs_mstvdv_fval_amount_ls.append(ele[sub_ele])
                elif sub_ele == 'isu_dcrs_mstvdv_amount': isu_dcrs_mstvdv_amount_ls.append(ele[sub_ele])
                
isu_dcrs_de_ls
```

```python
# DataFrame 만들기 https://eunguru.tistory.com/221
from pandas import Series, DataFrame
data = {
    '주식발행(감소)일자': isu_dcrs_de_ls,
    '발행(감소)형태': isu_dcrs_stle_ls,
    '주식종류': isu_dcrs_stock_knd_ls,
    '발행(감소)수량': isu_dcrs_qy_ls,
    '주당 액면가액': isu_dcrs_mstvdv_fval_amount_ls,
    '주당 발행(감소) 가액': isu_dcrs_mstvdv_amount_ls
}
df_irds = DataFrame(data)
df_irds = df_irds.sort_values(['주식발행(감소)일자'], ascending=True)
df_irds
```

### 2. 배당에 관한 사항
> __위치__
* 사업보고서 > I. 회사의 개요 > 6. 배당에 관한 사항 등

> __기능__
* 배당금총액, 배당수익률(%), 주당 배당금(원) 등 주요 배당 지표 알려줌

> __획득 가능 주요 정보__
* se : 배당금총액/배당수익률(%)/주당 배당금(원) 등 주요 배당 지표
* stock_knd : 보통주/우선주
* thstrm : 당기
* frmtrm : 전기
* lwfr : 전전기

```python
# 요청인자
crtfc_key = '27d86452b033fe973df6f1c980c37a2586801a1a'
corp_code = '00117212' # 두산
bsns_year = '2018'
reprt_code = '11011'
```

```python
url_basic = "https://opendart.fss.or.kr/api/alotMatter.json"
url_rest = "?crtfc_key={}&corp_code={}&bsns_year={}&reprt_code={}".format(crtfc_key, corp_code, bsns_year, reprt_code)
url = url_basic + url_rest
resp = requests.get(url)
webpage = resp.content.decode('UTF-8')
webpage
```

```python
se_ls = []
stock_knd_ls = []
thstrm_ls = []
frmtrm_ls = []
lwfr_ls = []

dict = json.loads(webpage)
for val in dict:
    if val == 'list':
        for i, ele in enumerate(dict[val]):
            for sub_ele in ele:
                if sub_ele == 'se': se_ls.append(ele[sub_ele])
                elif sub_ele == 'stock_knd': stock_knd_ls.append(ele[sub_ele])
                elif sub_ele == 'thstrm': thstrm_ls.append(ele[sub_ele])
                elif sub_ele == 'frmtrm': frmtrm_ls.append(ele[sub_ele])
                elif sub_ele == 'lwfr': lwfr_ls.append(ele[sub_ele])
                    
            if len(stock_knd_ls)-1 != i : stock_knd_ls.append('') 
        
                
stock_knd_ls
```

```python
data = {
    '구분': se_ls,
    '주식 종류': stock_knd_ls,
    '당기': thstrm_ls,
    '전기': frmtrm_ls,
    '전전기': lwfr_ls,
}
df_div = DataFrame(data)
df_div
```
