```python
import zipfile
import os
```

## 1. 압축하기
> 1.1. 개별파일 압축하기

```python
os.chdir('C:/05.Python/DartAPI/')

my_zip = zipfile.ZipFile("test_zip.zip", 'w') # zip파일, 쓰기모드
my_zip.write('pocket.csv')  # 압축할 파일
my_zip.close()
```

> 1.2. 여러파일 압축하기

```python
os.chdir('C:/05.Python/DartAPI/')

file_list = ['2018_ko.xls', 'pocket.csv']
with zipfile.ZipFile("test_zip.zip", 'w') as my_zip:
    for file in file_list:
        my_zip.write(file)
    my_zip.close()
```

## 2. 압축해제
> 2.1. 개별파일 압축해제: extract('압축해제하고 싶은 파일')

```python
zipfile.ZipFile('test_zip.zip').extract('pocket.csv')
```

> 2.2. 모든파일 압축해제: extractall()

```python
zipfile.ZipFile('test_zip.zip').extractall()
```

## 3. zip 파일 읽기
> 3.1. 특정 파일 읽기: read('읽고 싶은 파일명')

```python
import pandas as pd
import numpy as np
from io import BytesIO
```

```python
os.chdir('C:/05.Python/DartAPI/')
my_zip = zipfile.ZipFile('test_zip.zip')
file = my_zip.read('2018_ko.xls')
table = BytesIO(file)
pd.read_excel(table, sheet_name="연결 재무상태표", skiprows=6)
```

> 3.2. 압축파일 내 파일명 읽기: namelist()

```python
os.chdir('C:/05.Python/DartAPI/')
my_zip = zipfile.ZipFile('test_zip.zip')
my_zip.namelist()
```
